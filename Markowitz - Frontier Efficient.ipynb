


# FRONTIER EFFICIENT

```python
import numpy as np
import pandas as pd
import statsmodels as st
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ticker1 = 'EGIE3.SA'
ticker2 = 'PTBL3.SA'
ticker3 = 'VLID3.SA'
ticker4 = 'ITSA4.SA'
ticker5 = 'GGBR4.SA'

prices= pd.DataFrame()
tickers = [ticker1,ticker2,ticker3,ticker4,ticker5]
for t in tickers:
     prices[t]= wb.DataReader(t, data_source='yahoo', start='2010-1-1')['Adj Close']
prices.columns = ['EGIE3.SA','PTBL3.SA','VLID3.SA','ITSA4.SA','GGBR4.SA']
```

```python
prices
```

```python
log_ret= np.log(prices / prices.shift(1))
```

```python
#Pfolio random

np.random.seed(42)
num_ports = 10000
all_pesos = np.zeros((num_ports, len(prices.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    pesos = np.array(np.random.random(5))
    pesos = pesos / np.sum(pesos)
    all_pesos[x,:] = pesos
    ret_arr[x] = np.sum( (log_ret.mean() * pesos * 250))
    vol_arr[x] = np.sqrt(np.dot(pesos.T, np.dot(log_ret.cov()*250, pesos)))
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]
```

```python
#Pfolio real

all_pesos_r = np.zeros((num_ports, len(prices.columns)))
ret_arr_r = np.zeros(num_ports)
vol_arr_r = np.zeros(num_ports)
sharpe_arr_r = np.zeros(num_ports)

for p in range(num_ports):
    pesos_r = np.array([0.122504,0.291659,0.088584,0.299431,0.197822])
    all_pesos_r[p,:] = pesos_r
    ret_arr_r[p] = np.sum( (log_ret.mean() * pesos_r * 250))
    vol_arr_r[p] = np.sqrt(np.dot(pesos_r.T, np.dot(log_ret.cov()*250, pesos_r)))
    sharpe_arr_r[p] = ret_arr[p]/vol_arr[p]
```

```python
print("Max sharpe ratio in the array: {} ".format(sharpe_arr.max()))
print("Its location in the array: {} ".format(sharpe_arr.argmax()))
```

```python
print(all_pesos[7050,:])
max_sr_ret = ret_arr[sharpe_arr.argmax()]
max_sr_vol = vol_arr[sharpe_arr.argmax()]
max_sr_ret_r = ret_arr_r[sharpe_arr_r.argmax()]
max_sr_vol_r = vol_arr_r[sharpe_arr_r.argmax()]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_sr_vol, max_sr_ret,c='red', s=50) # red dot
plt.scatter(max_sr_vol_r, max_sr_ret_r, c= 'blue', s=50)
plt.show()
```
